# Classes

| $\epsilon_1$  | $\epsilon_2$   |Class|
|-----|------|-----|
| >0.5| >0.5 |  3  |
| >0.5| <0.5 |  2  |
| <0.5| >0.5 |  1  |
| <0.5| <0.5 |  0  |

```python
def get_euclidean_distance_from_boundary(predicted_class, e1, e2):
    if e1 < 0.5 and e2 < 0.5:
        actual_class = 0
    elif e1 < 0.5 and e2 > 0.5:
        actual_class = 1
    elif e1 > 0.5 and e2 < 0.5:
        actual_class = 2
    else:
        actual_class = 3
    eucliean_distance = math.sqrt((e1 - 0.5)**2 + (e2 - 0.5)**2)
    return predicted_class == actual_class, eucliean_distance
    
```

```python
get_euclidean_distance_from_boundary(2, 0.4, 0.8)
```

```python
#data['year'] = data['date'].apply(lambda x: x[0:4])
# df['col_3'] = df.apply(lambda x: f(x.col_1, x.col_2), axis=1)
df['euclidean_distance'] = df.apply(lambda x : get_euclidean_distance_from_boundary(x['class'], x['e_1'], x['e_2'])[1], axis=1)
```

```python
df['correct_prediction'] = df.apply(lambda x: x['class'] == x['predicted_class'], axis=1)
```
